## MNIST Using CNN

```python
# Import Dependencies
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# Load Data
mnist = input_data.read_data_sets('./dataset/MNIST_data/', one_hot=True)
```

### MNIST CNN Architecture

```python
from IPython.display import Image
Image('./cnnmnist.png')
```

```python
# Helper Functions

# Initialize Weights
def init_weights(shape):
    # Create Random Distribution with "Mean = 0" and "Standard Deviation = 0.1"
    init_random_dist = tf.truncated_normal(shape, mean=0, stddev=0.1)
    return tf.Variable(init_random_dist)
```

```python
# Initialize Bias
def init_bias(shape):
    # All bias values initialized to constant value = 0.1
    init_bias_vals = tf.constant(0.1,shape=shape)
    return tf.Variable(init_bias_vals)
```

```python
# CONV2D
def conv2D(x,W,stride=[1,1,1,1],pad='SAME'):
    # x: Input Tensor, shape = [batch, Height, Width, Channels]
    # W (Kernel): [filter Height, filter Width, Channels In, Channels Out]
    # 'SAME': Pad with 0's, 'VALID': 
    return tf.nn.conv2d(x, W, strides=stride, padding=pad)
```

```python
# Pooling Layer
def max_pooling(x, ks=[1,2,2,1], stride=[1,2,2,1], pad='SAME'):
    # "x" : Input Tensor, shape = [batch, Height, Width, Channels]
    # "ksize" : Size of Window for each dimension of I/P Tensor 
    return tf.nn.max_pool(x,ksize=ks, strides=stride, padding=pad)
```

```python
# Convolutional Layer
def convolutional_layer(input_x, shape):
    W = init_weights(shape)
    b = init_bias([shape[3]])
    return tf.nn.relu(conv2D(input_x, W) + b)
```

```python
# Fully Connected Layer
def fully_connected_layer(input_layer, size):
    input_size = int(input_layer.get_shape()[1])
    W = init_weights([input_size, size])
    b = init_bias([size])
    return tf.matmul(input_layer,W) + b
```

```python
# Build the CNN

# Placeholders
X = tf.placeholder(tf.float32, shape=[None,784])
y = tf.placeholder(tf.float32, shape=[None,10])
```

```python
# CNN Layers
# Reshape flattened values into image again use, 28*28
x_image = tf.reshape(X,[-1,28,28,1])
```

```python
# 1st Layer
# shape = [patch size, patch size, input channels, features]
# 32 features for each 5x5 patch
conv1 = convolutional_layer(x_image, shape=[5,5,1,32])
pool1 = max_pooling(conv1)
```

```python
# 2nd Layer
conv2 = convolutional_layer(pool1, shape=[5,5,32,64])
pool2 = max_pooling(conv2)
```

```python
# Flatten Layer
flatten = tf.reshape(pool2, [-1,7*7*64])
full_layer = tf.nn.relu(fully_connected_layer(flatten, 1024))
```

```python
# Dropout layer
hold_probb = tf.placeholder(tf.float32)
dropout = tf.nn.dropout(full_layer, keep_prob=hold_probb)
```

```python
y_pred = fully_connected_layer(dropout, 10)
```

```python
# Creat a Loss Function
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=y_pred))
```

```python
# Optimizer
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost) 
```

```python
init = tf.global_variables_initializer()
```

```python
steps = 1000

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(steps):
        batch_x, batch_y = mnist.train.next_batch(100)
        result, err = sess.run([optimizer,cost], feed_dict={X: batch_x, y: batch_y, hold_probb: 0.5})
        
        if i%100 == 0:
            matches = tf.equal(tf.argmax(y_pred,1), tf.argmax(y,1))
            acc = tf.reduce_mean(tf.cast(matches, tf.float32))
            print('Step: {0} , Accuracy: {1} %\n'.format(i, sess.run(acc, feed_dict={X: mnist.test.images, y: mnist.test.labels, hold_probb: 1.0}) * 100))
```
