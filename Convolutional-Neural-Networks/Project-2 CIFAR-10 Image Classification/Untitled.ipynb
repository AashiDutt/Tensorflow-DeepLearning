## Data Visualization

```python
# Generate Random Index Values for Plotting Training Images
rand_idx = np.random.randint(1,50000,10)
```

```python
rand_idx
```

```python
# Image Labels
labels = {0:'airplane', 1:'automobile', 2:'bird', 3:'cat', 4: 'deer', 5: 'dog', 6: 'frog', 7: 'horse', 8: 'ship', 9: 'truck'}
```

```python
# Plot Random Image
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15,15))
xcount = 0
ycount = 0
for idx in rand_idx:
    ax[xcount, ycount].imshow(X_train[idx])
    l = y_train[idx]
    n = l[0]
    ax[xcount, ycount].set_title(labels[n])
    
    if (ycount < 4):
        ycount +=1
    else:
        xcount += 1
        ycount = 0
```
